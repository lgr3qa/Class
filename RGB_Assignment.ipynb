```python
import requests
from PIL import Image
from io import BytesIO
import numpy as np
import matplotlib.pyplot as plt

def apply_red_filter(image_array):
    red_filtered = image_array.copy()
    red_filtered[:, :, 1:] = 0
    return red_filtered

def apply_green_filter(image_array):
    green_filtered = image_array.copy()
    green_filtered[:, :, [0, 2]] = 0
    return green_filtered

def apply_blue_filter(image_array):
    blue_filtered = image_array.copy()
    blue_filtered[:, :, :2] = 0
    return blue_filtered

def combine_channels(image_array, channels):
    combined = image_array.copy()
    for i in range(3):
        if i not in channels:
            combined[:, :, i] = 0
    return combined


url = input("Please enter the URL of a .jpg file: ")


response = requests.get(url)

if response.status_code == 200:
    original_image = Image.open(BytesIO(response.content))
    original_image_np = np.array(original_image)

    red_filtered = apply_red_filter(original_image_np)
    green_filtered = apply_green_filter(original_image_np)
    blue_filtered = apply_blue_filter(original_image_np)

    rg_filtered = combine_channels(original_image_np, [0, 1])  # Red and Green
    rb_filtered = combine_channels(original_image_np, [0, 2])  # Red and Blue
    gb_filtered = combine_channels(original_image_np, [1, 2])  # Green and Blue

    fig, axs = plt.subplots(3, 3, figsize=(12, 8))

    axs[0, 0].imshow(original_image)
    axs[0, 0].set_title('Original Image')
    axs[0, 0].axis('off')

    axs[0, 1].imshow(red_filtered)
    axs[0, 1].set_title('Red-Filtered Image')
    axs[0, 1].axis('off')

    axs[0, 2].imshow(green_filtered)
    axs[0, 2].set_title('Green-Filtered Image')
    axs[0, 2].axis('off')

    axs[1, 0].imshow(blue_filtered)
    axs[1, 0].set_title('Blue-Filtered Image')
    axs[1, 0].axis('off')

    axs[1, 1].imshow(rg_filtered)
    axs[1, 1].set_title('Red and Green Combined')
    axs[1, 1].axis('off')

    axs[1, 2].imshow(rb_filtered)
    axs[1, 2].set_title('Red and Blue Combined')
    axs[1, 2].axis('off')

    axs[2, 0].imshow(gb_filtered)
    axs[2, 0].set_title('Green and Blue Combined')
    axs[2, 0].axis('off')

    for i in range(1, 3):
        axs[2, i].axis('off')

    plt.tight_layout()
    plt.show()

else:
    print("Failed to retrieve the image.")

```
